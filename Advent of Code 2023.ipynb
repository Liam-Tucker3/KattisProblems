```python
# Creates a list of the text of each line of the file
def read_file(filename):
    lines = []
    data = open(filename, 'r')
    
    while(1):
        line = data.readline()
        if line: lines.append(line.strip())
        else: return lines
```

## Day 1

```python
# Day 1 Problem A
def first_last_dig_sum(filename):
    # Processing data
    lines = read_file(filename)
    
    score = 0
    for l in lines:
        nums = []
        for ch in l: 
            if ch.isdigit(): nums.append(ch)
        score += 10 * int(nums[0])
        score += int(nums[-1])
    
    return score

score = first_last_dig_sum('day01.txt')
print(score)
    
    
```

```python
# Day 1 Problem B
def first_last_dig_sum_B(filename):
    # Processing data
    lines = read_file(filename)
    
    score = 0
    for l in lines:
        nums = []
        for i in range(len(l)): 
            ch = l[i]
            if ch.isdigit(): nums.append(ch)
            if ch == 'o' and i+2 < len(l) and l[i:i+3] == 'one': nums.append('1')
            if ch == 't' and i+2 < len(l) and l[i:i+3] == 'two': nums.append('2')
            if ch == 't' and i+4 < len(l) and l[i:i+5] == 'three': nums.append('3')
            if ch == 'f' and i+3 < len(l) and l[i:i+4] == 'four': nums.append('4')
            if ch == 'f' and i+3 < len(l) and l[i:i+4] == 'five': nums.append('5')
            if ch == 's' and i+2 < len(l) and l[i:i+3] == 'six': nums.append('6')
            if ch == 's' and i+4 < len(l) and l[i:i+5] == 'seven': nums.append('7')
            if ch == 'e' and i+4 < len(l) and l[i:i+5] == 'eight': nums.append('8')
            if ch == 'n' and i+3 < len(l) and l[i:i+4] == 'nine': nums.append('9')
            
        score += 10 * int(nums[0])
        score += int(nums[-1])
    
    return score

score = first_last_dig_sum_B('day01.txt')
print(score)

```

```python
##### Day 2 Problem A

def legal_game(filename):
    lines = read_file(filename)
    
    score = 0
    for l in lines:
        
        passed = True
        game_index = int(l.split(':')[0][5:])
        this_line = l.split(':')[1].split(';')
        for draw in this_line:
            # print(draw)
            balls = draw.split(',')
            for b in balls:
                # print(b)
                vals = b.split(' ')
                for v in vals: 
                    if v == "": 
                        vals.remove(v)
                # print(vals)
                if int(vals[0]) > 12 and vals[1].strip() == "red" or int(vals[0]) > 13 and vals[1].strip() == "green" or int(vals[0]) > 14 and vals[1].strip() == "blue":
                    passed = False
        if passed: score += game_index
    return score
        
score = legal_game('day02.txt')
print(score)
        
```

```python
##### Day 2 Problem B

def sum_cube_power(filename):
    lines = read_file(filename)
    
    score = 0
    for l in lines:
        
        passed = True
        game_index = int(l.split(':')[0][5:])
        this_line = l.split(':')[1].split(';')
        
        min_red = 0
        min_blue = 0
        min_green = 0
        
        for draw in this_line:
            # print(draw)
            balls = draw.split(',')
            for b in balls:
                # print(b)
                vals = b.split(' ')
                for v in vals: 
                    if v == "": 
                        vals.remove(v)
                # print(vals)
                if vals[1].strip() == "red": min_red = max(int(vals[0]), min_red)
                if vals[1].strip() == "green": min_green = max(int(vals[0]), min_green)
                if vals[1].strip() == "blue": min_blue= max(int(vals[0]), min_blue)
                
        # print(min_red, min_blue, min_green)        
        score += min_red * min_blue * min_green
    
    return score
        
score = sum_cube_power('day02.txt')
print(score)
        
```

```python
def check_char(lines, i, j):
    if i-1 >= 0 and not lines[i-1][j].isdigit() and lines[i-1][j] != ".": return True
    if j-1 >= 0 and not lines[i][j-1].isdigit() and lines[i][j-1] != ".": return True

    if i+1 < len(lines) and not lines[i+1][j].isdigit() and lines[i+1][j] != ".": return True
    if j+1 < len(lines[0]) and not lines[i][j+1].isdigit() and lines[i][j+1] != ".": return True

    if i-1 >= 0 and j-1 >= 0 and not lines[i-1][j-1].isdigit() and lines[i-1][j-1] != ".": return True
    if i-1 >= 0 and j+1 < len(lines[0]) and not lines[i-1][j+1].isdigit() and lines[i-1][j+1] != ".": return True
    if i+1 < len(lines) and j+1 < len(lines[0]) and not lines[i+1][j+1].isdigit() and lines[i+1][j+1] != ".": return True
    if i+1 < len(lines) and j-1 >= 0 and not lines[i+1][j-1].isdigit() and lines[i+1][j-1] != ".": return True

    return False
    

def count_nums(filename):
    lines = read_file(filename)
    
    score = 0
    
    for i in range(len(lines)):
        this_val = 0
        is_num = False
        for j in range(len(lines[i])):
            if lines[i][j].isdigit():
                this_val = 10 * this_val + int(lines[i][j])
                if check_char(lines, i, j): is_num = True
                
            else:
                if is_num: score += this_val
                this_val = 0
                is_num = False
    
    return score

score = count_nums('day03test.txt')
print(score)
```

```python
# Day 04 Problem B
def score_card(str):
    nums = str.split(':')[1]
    vals = nums.split('|')
    wins = vals[0].split(' ')
    yours = vals[1].split(' ')
    
    score = 0
    d = {}
    for w in wins:
        d[w.strip()] = 0
    
    for y in yours:
        if y.strip() != "" and y.strip() in d: 
            score += 1
        
    return score

def total_cards(filename):
    lines = read_file(filename)
    card_counts = {}
    for i in range(len(lines)): card_counts[i+1] = 1
    
    for i in range(len(lines)):
        this_score = score_card(lines[i])
        for j in range(i+1, i+1+this_score):
            # print(i+1, card_counts[i+1])
            card_counts[j+1] = card_counts[j+1] + card_counts[i+1]
        
    total_cards = 0
    for k, v in card_counts.items():
        total_cards += v
        
    return total_cards

score = total_cards('day04.txt')
print(score)
```

```python
# Day 06 Problem A
def get_races(time_line, dist_line):
    d = {}
    time_vals = time_line.split()[1:]
    dist_vals = dist_line.split()[1:]
    for i in range(len(time_vals)):
        d[time_vals[i]] = dist_vals[i]
    return d
    
def calc_margin(filename):
    lines = read_file(filename)
    race_dict = get_races(lines[0], lines[1])
    
    cum_margin = 1
    for t, d in race_dict.items():
        
        # Iterating through every option. Not super efficient but fine for this test case
        this_num = 0
        t = int(t)
        d = int(d)
        for i in range(t):
            if d < (i * (t-i)): 
                this_num += 1
                # print(d, i, t-i, this_num)
        # print(this_num)
        cum_margin *= this_num
    
    return cum_margin
        
score = calc_margin('day06.txt')
print(score)
```

```python
# Day 07 Problem A
from functools import cmp_to_key

def get_rank(hand):
    d = {}
    for ch in hand:
        if ch in d.keys(): d[ch] = d[ch] + 1
        else: d[ch] = 1
    vals = list(reversed(sorted(d.values())))
    for i in range(len(vals)): vals[i] = str(vals[i])
    vals = ''.join(vals)
    
    if vals == "5": return 7
    elif vals == "41": return 6
    elif vals == "32": return 5
    elif vals == "311": return 4
    elif vals == "221": return 3
    elif vals == "2111": return 2
    elif vals == "11111": return 1
    else: return 0 # Should never happen

def compare_ch(card1, card2):
    dict = {'A': 13, 'K': 12, 'Q': 11, 'J': 10, 'T': 9, '9': 8, '8': 7, '7': 6, '6': 5, '5': 4, '4': 3, '3': 2, '2': 1}
    return dict[card1] - dict[card2]


def compare(item1, item2):
    score1 = get_rank(item1)
    score2 = get_rank(item2)
    if score1 > score2: 
        return -1
    if score1 < score2: 
        return 1
    else:
        assert score1 == score2
        for i in range(5):
            this_card_result = compare_ch(item1[i], item2[i])
            if this_card_result > 0: return -1
            if this_card_result < 0: return 1
    
    return 0
    
def total_score(filename):
    lines = read_file(filename)
    
    bid_lookup = {}
    hands = []
    hands2 = []
    
    for l in lines:
        vals = l.split()
        hand = vals[0]
        bid = int(vals[1])
        bid_lookup[hand] = bid
        hands.append(hand)
        hands2.append(str(get_rank(hand)) + hand)
        
    hands = list(reversed(sorted(hands, key=cmp_to_key(compare))))
    # print(hands)
    
    score = 0
    for i in range(len(hands)):
        score += (i+1) * bid_lookup[hands[i]]
        
    return score
    
score = total_score('day07.txt')
print(score)
```

```python
# Day 07 Problem B
from functools import cmp_to_key

def get_rank(hand):
    d = {}
    for ch in hand:
        if ch in d.keys(): d[ch] = d[ch] + 1
        else: d[ch] = 1
        
    j_val = 0
    if "J" in d.keys():
        j_val = d["J"]
        del d["J"]
        
    if j_val == 5: return 7 # For all J case
    
    vals = list(reversed(sorted(d.values())))
    vals[0] += j_val
    for i in range(len(vals)): vals[i] = str(vals[i])
    vals = ''.join(vals)
    
    if vals == "5": return 7
    elif vals == "41": return 6
    elif vals == "32": return 5
    elif vals == "311": return 4
    elif vals == "221": return 3
    elif vals == "2111": return 2
    elif vals == "11111": return 1
    else: return 0 # Should never happen

def compare_ch(card1, card2):
    dict = {'A': 13, 'K': 12, 'Q': 11, 'J': 0, 'T': 9, '9': 8, '8': 7, '7': 6, '6': 5, '5': 4, '4': 3, '3': 2, '2': 1}
    return dict[card1] - dict[card2]


def compare(item1, item2):
    score1 = get_rank(item1)
    score2 = get_rank(item2)
    if score1 > score2: 
        return -1
    if score1 < score2: 
        return 1
    else:
        assert score1 == score2
        for i in range(5):
            this_card_result = compare_ch(item1[i], item2[i])
            if this_card_result > 0: return -1
            if this_card_result < 0: return 1
    
    return 0
    
def total_score(filename):
    lines = read_file(filename)
    
    bid_lookup = {}
    hands = []
    hands2 = []
    
    for l in lines:
        vals = l.split()
        hand = vals[0]
        bid = int(vals[1])
        bid_lookup[hand] = bid
        hands.append(hand)
        hands2.append(str(get_rank(hand)) + hand)
        
    hands = list(reversed(sorted(hands, key=cmp_to_key(compare))))
    # print(hands)
    
    score = 0
    for i in range(len(hands)):
        score += (i+1) * bid_lookup[hands[i]]
        
    return score
    
score = total_score('day07.txt')
print(score)
```

## Day 08

```python
def get_dist(filename):
    lines = read_file(filename)
    
    left_dict = {}
    right_dict = {}
    
    path = lines[0].strip()
    print(path)
    lines = lines[2:]
    
    curr_locs = []
    
    for l in lines:
        vals = l.split()
        source = vals[0]
        left = vals[2][1:4]
        right = vals[3][:3]
        
        left_dict[source] = left
        right_dict[source] = right
        
        if source[2] == "A": curr_locs.append(source)
        
    print("Number of Ghosts:", len(curr_locs))
        
    steps = 0
    while True:
        if steps % 250000 == 0:
            print(steps)
        
        if path[steps % len(path)] == "L":
            for i in range(len(curr_locs)):
                curr_locs[i] = left_dict[curr_locs[i]]
        else:
            for i in range(len(curr_locs)):
                curr_locs[i] = right_dict[curr_locs[i]]
        steps += 1
        
        all_xxz = True
        for c in curr_locs:
            if c[2] != "Z": 
                all_xxz = False
                break
        if all_xxz: return steps
    
    
score = get_dist('day08.txt')
print(score)
```

```python
# Day 09 Problem A
def next_val_rec(l):
    all_zero = True
    for num in l:
        if l != 0: all_zero = False
        
    # Base Case
    if all_zero: return 0

    else:
        new_l = []
        # print(l)
        for i in range(len(l) - 1):
            new_l.append(int(l[i+1]) - int(l[i]))
        num = next_val_rec(new_l)
        # print(len(l))
        return l[-1] + num

def sum_extrap_val(filename):
    lines = read_file(filename)
    
    score = 0
    for l in lines:
        vals = l.split()
        for i in range(len(vals)):
            vals[i] = int(vals[i])
        # print(vals)
        score += next_val_rec(vals)
    
    return score

score = sum_extrap_val('day09.txt')
print(score)
```

```python
# Day 09 Problem B
def last_val_rec(l):
    all_zero = True
    for num in l:
        if l != 0: all_zero = False
        
    # Base Case
    if all_zero: return 100

    else:
        new_l = []
        # print(l)
        for i in range(len(l) - 1):
            new_l.append(int(l[i+1]) - int(l[i]))
        num = next_val_rec(new_l)
        # print(len(l))
        return num + l[-1]

def sum_extrap_val(filename):
    lines = read_file(filename)
    
    score = 0
    for l in lines:
        vals = l.split()
        for i in range(len(vals)):
            vals[i] = int(vals[i])
        vals = list(reversed(vals))
        # print(vals)
        score += next_val_rec(vals)
    
    return score

score = sum_extrap_val('day09.txt')
print(score)
```

```python
# Day 10 Problem A
def get_len_rec(grid, count, i, j, old_dir):
    print(count)
    
    # Base cases
    if i < 0 or i >= len(grid) or j < 0 or j >= len(grid[i]): return -1
    # print(i, j, grid[i][j])

    if grid[i][j] == "S": return count # Increment when visiting new location
    if grid[i][j] == ".": return -1

    this_val = grid[i][j]
    grid[i] = grid[i][:j] + "X" + grid[i][j+1:]

    # Case 1: coming from the bottom:
    if old_dir == "S":
        if this_val == "|": return get_len_rec(grid, count+1, i-1, j, "S")
        if this_val == "7": return get_len_rec(grid, count+1, i, j-1, "R")
        if this_val == "F": return get_len_rec(grid, count+1, i, j+1, "L")

    if old_dir == "N":
        if this_val == "|": return get_len_rec(grid, count+1, i+1, j, "N")
        if this_val == "L": return get_len_rec(grid, count+1, i, j+1, "L")
        if this_val == "J": return get_len_rec(grid, count+1, i, j-1, "R")
    
    if old_dir == "R":
        if this_val == "-": return get_len_rec(grid, count+1, i, j-1, "R")
        if this_val == "L": return get_len_rec(grid, count+1, i-1, j, "S")
        if this_val == "F": return get_len_rec(grid, count+1, i+1, j, "N")
    
    if old_dir == "L":
        if this_val == "-": return get_len_rec(grid, count+1, i, j+1, "L")
        if this_val == "7": return get_len_rec(grid, count+1, i+1, j, "N")
        if this_val == "J": return get_len_rec(grid, count+1, i-1, j, "S")
    
    return -2

def get_len(filename):
    lines = read_file(filename)
    lines1 = read_file(filename)
    lines2 = read_file(filename)
    lines3 = read_file(filename)
    lines4 = read_file(filename)
    # print(lines)
    
    # Getting index of "S"
    r = -1
    c = -1
    for i in range(len(lines)):
        for j in range(len(lines[0])):
            if lines[i][j] == "S":
                r = i
                c = j
                
    # print(r, c)
    
    s1, s2, s3, s4 = 0, 0, 0, 0
    print("A")
    if c-1 > 0: s3 = get_len_rec(lines1, 1, r, c-1, "R")
    print("B")
    if c+1 > 0: s4 = get_len_rec(lines2, 1, r, c+1, "L")
    print("C")
    if r-1 > 0: s1 = get_len_rec(lines3, 1, r-1, c, "S")
    print("D")
    if r+1 > 0: s2 = get_len_rec(lines4, 1, r+1, c, "N")
    print("E")
    
    return int(max(s1, s2, s3, s4) / 2)

score = get_len("day10.txt")
print(score)
    
```
